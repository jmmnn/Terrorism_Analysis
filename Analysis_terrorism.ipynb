```python
import pandas
```

```python
'''
Configuration:
'''

#datafile = 'globalterrorismdb_0616dist.xlsx'
datafile = 'test.xlsx'

#Read data
myfile = pandas.ExcelFile(datafile)
#myfile.sheet_names #use this if the file has multiple sheets, get the name of the sheet you want to read.

#create the dataframe
mydataframe = myfile.parse('Data')
```

```python
#Plots
#Importing the package
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

#test that everything is working
print __version__ # requires version >= 1.9.0

# This is to render the plots on this notebook
# If you want the plots standalone, comment the line below:
init_notebook_mode(connected=True)
```

```python
#sample scatterplot
from plotly.graph_objs import Scatter, Figure, Layout
iplot([Scatter(x=mydataframe.country_txt, y=mydataframe.nkill)])
```

```python
#sample bubble charts

from plotly.graph_objs import *

iplot({
    'data': [
        Scatter(x=mydataframe.country_txt,
                y=mydataframe.iyear,
                text=mydataframe.country_txt)
    ],
    'layout': Layout(xaxis=XAxis(title='Life Expectancy'), yaxis=YAxis(title='GDP per Capita', type='log'))
}, show_link=False)
```
